```python
plt.figure(figsize=(10,6))
sns.boxplot(data=dt)
plt.xticks(rotation=90)
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode Interests into numerical values
label_encoder = LabelEncoder()
dt["tour duration"] = label_encoder.fit_transform(dt["Preferred Tour Duration"])

# Display unique encoded values
print(dt[["Preferred Tour Duration", "tour duration"]].drop_duplicates())
```

```python
correlation = dt["Age"].corr(dt["tour duration"])
print(f"Correlation between Age and Interests: {correlation}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter Plot
plt.figure(figsize=(8,5))
sns.scatterplot(x=dt["Age"], y=dt["Interests_Encoded"])
plt.title("Age vs Interests Correlation")
plt.xlabel("Age")
plt.ylabel("Interests (Encoded)")
plt.show()

# Heatmap
plt.figure(figsize=(5,3))
sns.heatmap(dt[["Age", "Interests_Encoded"]].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
